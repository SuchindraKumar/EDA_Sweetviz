```python
import pandas as pd
```

```python
import sweetviz

train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.head()
```

```python
test.head()
```

### Analyzing a Single Dataframe (and its optional target feature)

```python
my_report = sweetviz.analyze([train, "Train"],target_feat='SalePrice')
```

### Generate the report in html formate

```python
my_report.show_html('Train_Report.html')
```

### Comparing two dataframes (e.g. Test vs Training sets)

```python
combined_report = sweetviz.compare([train, "Train"], [test, "Test"], "SalePrice")
```

```python
combined_report.show_html('Combined_Report.html')
```
